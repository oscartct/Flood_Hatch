```python
import rasterio
import matplotlib.pyplot as plt

with rasterio.open('data/processed/flood_mask.tif') as src:
    mask = src.read(1)

# Plot with a stronger contrast
plt.figure(figsize=(6,6))
plt.imshow(mask, vmin=0, vmax=1)  # maps 0->black, 1->white
plt.title("Flood Mask (white = flooded)")
plt.axis('off')
plt.show()
```
